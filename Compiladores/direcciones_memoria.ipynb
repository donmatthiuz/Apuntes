# Direcciones de memoria en código destino

## Instrucciones

Analiza el siguiente fragmento de código de tres direcciones y tradúcelo a código ensamblador utilizando:

- Asignación estática.
- Asignación de pila.


```plaintext
call p
call q
return
call r
return
return
```



### Asignacion estatica

En nuestro main se veria algo asi

```plaintext
100: ST 364, #120
112: BR 200  // llamamos a p
120: ST 364, #140  
132: BR 300  // llamamos a 1
140: BR *364 // hace el return 
152: ST 364, #172 
164: BR 400  // llama a r
172: BR *364 // return inicial 
184: BR *364  // return final
 
```
Ahora en p se veria asi

```plaintext
200: ST 364, #220   // guarda dirección de retorno antes de llamar a q
212: BR 300         // call q
220: BR *364        // return

 
```

Ahora en q se veria asi

```plaintext
300: ACTION ...
320: BR *364        // return
```

Ahora r se veria asi

```plaintext
400: ACTION ...
420: BR *364        // return
```


### Asignacion de Pila

El codigo principal seria 

```plaintext
100: ST 500, #600        // inicializa el puntero de pila (SP = 600)
110: ST *500, #120        // guarda dirección de retorno (push)
112: ADD 500, #1          // SP++
114: BR 200               // call p

120: ST *500, #140        // push return dir para q
122: ADD 500, #1
124: BR 300               // call q

140: ST *500, #160        // push return dir para r
142: ADD 500, #1
144: BR 400               // call r

160: HALT
```


El procedimiento de p seria

```plaintext
200: ST *500, #220        // push dirección de retorno
202: ADD 500, #1
204: BR 300               // call q

220: SUB 500, #1          // SP--
222: BR **500             // pop y return
```

El procedimiento de q seria

```plaintext
300: ACTION ...
310: SUB 500, #1          // SP--
312: BR **500             // return
```


El procedimiento de r seria

```plaintext
400: ACTION ...
410: SUB 500, #1          // SP--
412: BR **500             // return
```
